```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('sneaker_ad_spend.csv')
```

```python
df
```

```python
df.plot(kind='scatter',x='Instagram_Ad_Spend',y='Total_Sales')
plt.show()
```

```python
X=df[['Instagram_Ad_Spend']]
y=df['Total_Sales']
model=LinearRegression()
model.fit(X,y)
print("Model Training SUCCESSFUL")
```

```python
future_spend=pd.DataFrame({'Instagram_Ad_Spend':[1500]})
predicted_sales=model.predict(future_spend)
print("Predicted Sales:",predicted_sales.astype(int))
```

```python
accuracy=model.score(X,y)
print('Accuracy',accuracy*100)
```

```python
plt.scatter(X,y,color='Blue',label='Actual Historical Sales')
plt.plot(X,model.predict(X),linewidth=3,color='Red',label='Predicted Data')
plt.title('SneakerHub: Ad Spend vs. Sales Forecast')
plt.xlabel('Instagram Ad Spend ($)')
plt.ylabel('Total Sales ($)')
plt.legend()
plt.show()
```
